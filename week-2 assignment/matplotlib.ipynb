```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

basic graph

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [1,2,3]
y = [2,4,6]

# resize the graph
plt.figure(figsize=(5,3), dpi=300)
plt.plot(x,y, label = '2x', color='green', linewidth=2, marker='.',linestyle='--', markersize=10, markeredgecolor = 'blue')

x2 = np.arange(0,5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'r', label='x^2')
plt.plot(x2[4:],x2[4:]**2, 'y--', label='x^2')

plt.title("first graph", fontdict={'fontname':'comic sans ms','fontsize':20})
plt.xlabel("x axis")
plt.ylabel("y axis")

# can change the ticks which are visible
# plt.xticks([0,1,2,3,4])
# plt.yticks([0,1,3,4])

# legend marks the label to graphs, and to use label need to be given above
plt.legend()

# save the figure withe the high resolution
# plt.savefig('my graph.png', dpi = 300)

plt.show()
```

### bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,5]

bars = plt.bar(labels,values)
 # design in the graphs
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
# above thing can be done by the for loop too

patterns = ['/', 'o', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head(5)
```

```python
bins = [0,20,40,60,80,100]

plt.hist(fifa.Overall, bins=bins, color='orange')

plt.xticks(bins)

plt.ylabel("number of players")
plt.ylabel("skill level")
plt.title("distribution fo the player vs their skill in FIFA")

plt.show()
```

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']
plt.pie([left,right], labels = labels, colors=colors, autopct='%.3f %%')

plt.title("foot preference fo the fifa players")
plt.show()
```

```python
fifa.Weight.head(10)
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[(fifa.Weight < 125)].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Above 200']

explode = (.4,.1,0,0.05,.4)

plt.pie(weights, labels = labels, autopct = '%.3f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels= labels, patch_artist=True, medianprops={'linewidth':3})

colors
for box in boxes['boxes']:
    box.set(color="pink", linewidth=5)
    box.set(facecolor="grey")

plt.title("PROFESSIONAL SOCCER TEAM COMPARISON")
plt.ylabel('FIFA overall rating')

plt.show()
```
