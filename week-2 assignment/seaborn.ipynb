```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
# distribution plot

sns.distplot(crash_df['not_distracted'], kde=False, bins=25)

# kde is the line over the plots
```

```python
#joint plot

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

# we can make kde as the 'hex' or 'kde'
```

```python
# kde plot
sns.kdeplot(crash_df['alcohol'])
```

```python
# pair plots , plot the relationship

sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df

sns.pairplot(tips_df, hue='sex', palette='Reds') # reds and not red
```

```python
# rug plots, not used much

sns.rugplot(tips_df['tip'])
```

```python
# styling of the plots

sns.set_style('ticks')

sns.set_context('poster', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', color='pink', data=crash_df, kind='hex')

sns.despine(left=True) # this removes the axis of that side, even top nd bottom too
```

#### categorical plots

```python
# abar plots

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.var)

# we can use np.cov, np.std_dev and all
```

```python
# count plot
sns.countplot(x='sex',data=tips_df)
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
```

```python
# violin plot

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=False)
```

```python
# strip plot
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=False, hue='sex', dodge=False)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, color='brown')
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='pink')
```

```python
# pallette

plt.figure(figsize=(8,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')

plt.legend(loc=3)  # adjust the position of the box 
```

```python
# matrix plots or heat maps
# need the data in the matrix form

plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.5)

# here the data type needed to be converted
numeric_columns = crash_df.select_dtypes(include=[np.number])
crash_mx = numeric_columns.corr()

sns.heatmap(crash_mx, annot=True, cmap='Reds')  # first capital, reds give error
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights, cmap='Blues', linecolor='green', linewidths=0.5)
```

```python
# cluster maps

iris = sns.load_dataset('iris')
iris
```

```python
species = iris.pop('species')
sns.clustermap(iris)
```

```python
sns.clustermap(flights, cmap = 'Greens', standard_scale=1)
```

```python
# pairs grids, diff form pair plots

iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue="species")

iris_g.map(plt.scatter)
# histogram along diagonal
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g = sns.PairGrid(iris, hue='species',x_vars=['sepal_length', 'sepal_width'],
y_vars=['petal_length', 'petal_width'])

iris_g.map(plt.scatter)
iris_g.add_legend()
```

```python
# facet grid

tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.5)

tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```
